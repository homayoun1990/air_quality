#### Interpolation for creating air pollution map 

```python
import numpy as np
import pandas as pd
import glob
from pykrige.ok import OrdinaryKriging
from pykrige.kriging_tools import write_asc_grid
import pykrige.kriging_tools as kt
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.patches import Path, PathPatch
```

```python
header = ''' 
SELECT *
FROM Measurement_summary
LIMIT 1
'''
header = pd.read_sql_query(header, conn)
print ("Measurement_summary")
header
```

```python

type_of_columns = '''
PRAGMA table_info(Measurement_summary); 
         '''
d_type = pd.read_sql_query(type_of_columns, conn)
d_type
```

```python


lons = np.array(df_measurement['field5']) 
lats = np.array(df_measurement['field4'])

SO2 = np.array(df_measurement["field6"])
NO2 = np.array(df_measurement["field7"])
O3 = np.array(df_measurement["field8"])
CO = np.array(df_measurement["field9"])
PM10 = np.array(df_measurement["field10"])
PM25 = np.array(df_measurement["field11"])
```

```python
SO2
```

```python

grid_space = 0.001
grid_lon = np.arange(np.amin(lons), np.amax(lons), grid_space)
grid_lat = np.arange(np.amin(lats), np.amax(lats), grid_space)
```

```python
OK = OrdinaryKriging(lons, lats, NO2, variogram_model='gaussian', # linear or gaussian
                     verbose=True, enable_plotting=True, nlags=20)

z1, ss1 = OK.execute('grid', grid_lon, grid_lat)

plt.imshow(z1, origin="lower")
plt.show()
```
